Data Loading and Preparation

```python
import pandas as pd

data = pd.read_csv('./4K_House_Rent_Dataset.csv')
X = data['Size']
Y = data['Rent']
```

Loss Function

```python
import numpy as np

def loss(a, b):
    return 1/len(X) * np.sum((a*X + b - Y)**2)
```

Contour Plot

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(7, 5))
a_vals = np.arange(-400, 400, 5)
b_vals = np.arange(-300000, 300000, 20000)
A, B = np.meshgrid(a_vals, b_vals)
loss_vals = np.array([loss(a, b) for a, b in zip(np.ravel(A), np.ravel(B))]).reshape(A.shape)
contour = ax.contour(A, B, loss_vals, levels=30, cmap="viridis")
ax.set_xlabel('a')
ax.set_ylabel('b')
ax.set_title('Contour Plot')
plt.tick_params(axis='both', which='major', labelsize=7)
plt.show()
```

3D Surface Plot

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(7, 5))
ax = fig.add_subplot(111, projection='3d')
a_vals_3d = np.arange(-100, 200, 5)
b_vals_3d = np.arange(-180000, 150000, 15000)
A_3d, B_3d = np.meshgrid(a_vals_3d, b_vals_3d)
loss_vals_3d = np.array([loss(a, b) for a, b in zip(np.ravel(A_3d), np.ravel(B_3d))]).reshape(A_3d.shape)
ax.plot_surface(A_3d, B_3d, loss_vals_3d, cmap="viridis", rstride=3, cstride=3, edgecolor='none', alpha=0.8)
ax.contour(A_3d, B_3d, loss_vals_3d, levels=20, cmap="viridis", linestyles="solid", offset=0)
ax.view_init(40, 290)
ax.set_xlabel('a')
ax.set_ylabel('b')
ax.set_zlabel('loss', rotation="vertical")
ax.set_title('3D Surface Plot')
plt.tick_params(axis='both', which='major', labelsize=7)
plt.show()
```
